# 식별 연산자 (Identity Operators)
- 식별 연산자는 두 객체가 같은 객체인지를 확인한다.
- is : 두 객체가 같은 객체인지 확인
- is not 두 객체가 다른 객체인지 확인
- 결과 : True, False


````python
a = [1, 2, 3]
b = a
c = [1, 2, 3]
print(a is b)  # True (a와 b는 같은 객체)
print(a is c)  # False (a와 c는 같은 값을 가지지만 다른 객체)
````

#   멤버 연산자 (Membership Operators)
- 멤버 연산자는 시퀀스 내에 값이 존재하는지 확인한다.
- in : 특정값이 멤버의 값인지 확인
- not in : 특정값이 멤버의 값인지 아닌지 확인
- 결과 : True, False

````python
a = [1, 2, 3]
print(2 in a)  # True
print(4 not in a)  # True
````

## 반복문 : while
반복문에서 else는 반복문이 완벽하게 끝났을 때 실행된다.  
대부분의 프로그래밍 언어는 while과 for을 지원한다.  

else
  - if ~ else    
  - while ~ else  
  - try ~ except ~ else ~ finally


```
while 조건식:
  수행할 문장
else:
  수행할 문장
```


````python
# while문이 완벽이  돌았으므로 end가 출력된다.
a = 10
while  a > 5 :
  a -= 1  # a = a - 1
  print(a)
else:
  print('end')
````

````python
# while문을 강제적으로 빠져나왔기 때문에 else영역을 실행하지 않는다.
a = 10
while  a > 5 :
  a -= 1  # a = a - 1
  print(a)
  if a==5:
    break   # 반복문을 강제적으로 빠져나올때 사용한다. 즉 반복문을 중단시킨다
else:
  print('end')
````

````python
#  continue는 다음 루프(조건식)로 바로 넘어간다.
# 반복문을 중단하지는 않으므로 else문이 실행된다.
a = 10 #5
while a > 5:
  a -= 1
  if a % 2 == 0 :
    continue   # 이동
  print(a)
else:
  print('end')
````

# range

````python
#range(stop) # 0 ~ stop값 미만
for i in range(5):
    print(i)

print("\n")

#range(start, stop) # start ~ stop값 미만
for i in range(1, 6):
    print(i)

print("\n")

#range(start, stop, step) # start ~ stop값 미만
for i in range(1, 6, 2): # 1 ~ 6 미만 값이 2씩 증가해서 생성된다 (1, 3, 5)
    print(i)
````

## enumerate 함수
 - enumerate()함수는 인덱스와 객체를 동시에 저장할 수 있는 특징을 가진 for문을 좀더 효율적으로 사용하기 위한 함수이다.
 - enumerate()함수는 반복가능객체(list, tuple, str)을 입력으로 받아 인덱스를 포함하는 enumerate 객체를 리턴한다.
 - enumerate([순서자료형], [시작값=0])의 형식을 가지며 반복가능객체은 반복자가 가능한 객체가 입력되고 시작값은 입력시 해당수 부터 증가하며, 생략했을 경우 기본값은 0 이다.


````python
for k in enumerate(['apple', 'python', 'microsoft']):
    print(k)

print('\n')

for k, v in enumerate(['apple', 'python', 'microsoft']):
    print(k, v)

print('\n')

# enumerate(iterable, start) => start값은 생략 가능 default == 0
for k, v in enumerate(['apple', 'python', 'microsoft'], 1):
    print(k, v)

print('\n')
for k in enumerate('abcd'):
    print(k)
````

````python
a = {'one':10, 'two':20}
for k in enumerate(a):
  print(k)

print('\n')

a = {'one':10, 'two':20}
for i, k in enumerate(a):
  print(i, a[k])

print('\n')
#키 값만 받고 싶으면 인덱스 자리 _언더바 넣기
a = {'one':10, 'two':20}
for _, k in enumerate(a):
  print(a[k])
````
