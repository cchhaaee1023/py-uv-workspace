 [문제10]  
 if~else문을 이용해서 년도에 해당하는 띠를 구하는 프로그램을 구현하시오.  
 단, 년도은 input()으로 입력받음.

   1. 12가지 띠  
      원숭이, 닭, 개, 돼지, 쥐, 소, 호랑이, 토끼, 용, 뱀, 말, 양  
   
   2. 띠구하는 공식  
     year%12의 결과가  
     0이면 "원숭이", 1이면 "닭"  2이면 "개",    3이면 "돼지",  
     4이면 "쥐",    5이면 "소", 6이면 "호랑이", 7이면 "토끼",  
     8이면 "용",    9이면 "뱀", 10이면 "말",   11이면 "양"  
     
   3. 출력결과  
     2012년도의 태생은  용띠입니다.  

```python
year = int(input('년도를 입력: '))

if year % 12 == 0:
    re = '원숭이'
elif year % 12 == 1:
    re = '닭'
elif year % 12 == 2:
    re = '개'
elif year % 12 == 3:
    re = '돼지'
elif year % 12 == 4:
    re = '쥐'
elif year % 12 == 5:
    re = '소'
elif year % 12 == 6:
    re = '호랑이'
elif year % 12 == 7:
    re = '토끼'
elif year % 12 == 8:
    re = '용'
elif year % 12 == 9:
    re = '뱀'
elif year % 12 == 10:
    re = '말'
else:
    re = '양'

print(f'{year}년도의 태생은 {re}띠입니다.')
```

[문제12]
A 학급에 총 10명의 학생이 있다. 이 학생들의 중간고사 점수는 다음과 같다.

[70, 60, 55, 75, 95, 90, 80, 80, 85, 100]

list comprehension을 사용하여 A 학급의 평균 점수를 구해 보자.

```python
scores = [70, 60, 55, 75, 95, 90, 80, 80, 85, 100]

total = sum([i for i in scores])
print(f"average is {total / 10}")
```

[문제13]
======음식점 메뉴 관리 프로그램======
1. 신규 메뉴 등록
2. 메뉴판 전체 보기
3. 프로그램 종료
> 1
메뉴명: 짜짜
가격: 3000
신규 메뉴 짜짜(이)가 등록되었습니다.
{'짜짜': '3000'}

======음식점 메뉴 관리 프로그램======
1. 신규 메뉴 등록
2. 메뉴판 전체 보기
3. 프로그램 종료
> 2

==========메뉴판==========
짜짜 : 3000원
----------------------------------
1. 수정 | 2. 삭제 | 3. 나가기
> 3

======음식점 메뉴 관리 프로그램======
1. 신규 메뉴 등록
2. 메뉴판 전체 보기
3. 프로그램 종료
> 3
프로그램을 종료합니다.

```python
num = 0
menu = {}

while(num != 3):
    print('======음식점 메뉴 관리 프로그램======')
    print('1. 신규 메뉴 등록')
    print('2. 메뉴판 전체 보기')
    print('3. 프로그램 종료')

    num = int(input('> '))

    if num == 1:
        mname = str(input("메뉴 이름을 알려주세요: "))
        mprice = int(input("가격을 알려주세요: "))
        menu = {mname : mprice}
        print(f"신규 메뉴 {mname}(이)가 등록되었습니다.")
        
    elif num ==2:
        if menu:
            print("==========메뉴판==========")
            for i in menu:
                print(f"{i}: {menu[i]}원")
            print('----------------------------------')
            num2 = int(input('1. 수정 | 2. 삭제 | 3. 나가기: '))
            if num2 == 1:
                mename = str(input("수정할 메뉴명을 입력하세요: "))
                newname = str(input('수정된 이름을 입력하세요: '))
                newprice = int(input('수정된 가격을 입력하세요: '))
                menu[newname] = menu.pop(mename)
                menu[newname] = newprice
            elif num2 == 2:
                if menu.__len__ == 0:
                    print("메뉴가 없습니다.")
                else:
                    mename = str(input("삭제할 메뉴명을 입력하세요: "))
                    del menu[mename]
            elif num2 == 3:
                continue
        else:
            print("등록된 메뉴가 없습니다.")
        
    elif num == 3:
        print("프로그램 종료합니다.")
        break
```
