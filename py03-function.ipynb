## return
  - return의 정의

외부 함수의 자원을 내부 함수에서 참조는 가능하지만 변경은 불가능하다.
- 변경시에는 **nonlocal** 사용한다.

````python
k = 1
def outer():
    m = 1
    def inner():
        global k
        k = k + 1

        nonlocal m
        m = m + 1

        return k, m
    return inner()
````

````python
print(outer())
````

sorted() 정렬 기준(key) 지정
-데이터 정렬 시 key 옵션 활용
  - 딕셔너리나 튜플로 이루어진 복잡한 리스트를 정렬할 때 기준점을 지정

````python
students = [('Kim', 85), ('Lee', 92), ('Park', 78)]

#점수 기준 오름차순 정렬 => reverse파라미터의 default = False
proc_sorted = sorted(students, key=lambda x : x[1])
print(proc_sorted)

#점수를 기준으로 내림차순 정렬
proc_sorted = sorted(students, key = lambda x : x[1], reverse = True) #x[0]이면 이름 가지고 정렬 # reverse 내림차순
print(proc_sorted)
````

## 재귀함수


```
재귀함수란?  
  - 함수 내부에서 자기 자신을 다시 호출하는 구조를 가진 함수  
  - 복잡한 문제를 더 작은 단위의 동일한 문제로 나누어 해결할 때 유용 (분할정복)


 재귀함수의 필수 2대 요소  
  - 1. 기저 조건 (Base Case): 재귀 호출을 멈추고 탈출하는 조건 (구현 누락 시 무한 루프)
  - 2. 재귀 단계 (Recursive Case): 문제를 작게 쪼개어 자기 자신을 다시 호출하는 부분


 기본 개념 구조  
   def proc(n):      
        if n <= 1:                    # 1. 기저 조건          
               return 1      
        return n + proc(n-1)     # 2. 재귀 단계 (자기 자신 호출)


```

### 팩토리얼 (n!) 구현
  - n부터 1까지의 모든 양의 정수를 곱하는 연산 (n! = n * (n-1) * (n-2) * ... * 1)

````python
def proc_factorial(n):
    if n <= 1:
        return 1
    else:
        return n * proc_factorial(n-1)
````

````python
print(proc_factorial(5))
````

````python
# 반복문으로 팩토리얼 구하기
def proc_factorial(n):
      result = 1
      # 1부터 n까지 순차적으로 곱하기 진행
      for i in range(1, n + 1):
          result *= i
      return result


# 함수 호출 및 결과 확인
print(proc_factorial(5))  # → 출력: 120 (1 * 2 * 3 * 4 * 5)
````

### 메모이제이션(Memoization)의 개념과 특징
```
①메모이제이션(Memoization) 기술이란?
  - 컴퓨터 프로그램이 동일한 계산을 반복해야 할 때, 이전에 계산한 값을
    메모리에 저장(캐싱)함으로써 동일한 중복 연산을 제거하는 최적화 기법입니다.


② 핵심 작동 원리 2가지
  - 결과 값 캐싱: 한 번 계산한 결과는 딕셔너리(Dictionary) 등의 메모리 공간에 기록
  - 재귀 호출 스킵: 이미 계산된 값이 필요할 때는 재귀를 수행하지 않고 즉시 꺼내어 재사용


③ 도입 시 기대 효과
  - 무수히 반복되는 중복 연산 수치를 제로(0)에 가깝게 만듦
  - 시간 복잡도를 혁신적으로 줄여 프로그램의 실행 속도를 기하급수적으로 향상시킴
```

````python
#memo
memo = {}

def memofact(n):
    if n <= 1:
        return 1
    if n in memo:
        return memo[n]
    
    memo[n] = n * memofact(n-1)

    return memo[n]


print(memofact(6))
````

### 파이썬 파이토닉(Pythonic) 메모이제이션 데코레이터
-   @lru_cache 데코레이터 활용
  - 파이썬 내장 라이브러리를 사용하면 코드 수정 없이 한 줄로 자동 메모리화 가능

````python
from functools import lru_cache

# maxsize=None은 제한 없이 모든 결과를 메모리에 캐싱하겠다는 의미
@lru_cache(maxsize=None)
def proc_factorial(n):
      if n <= 1:
          return 1
      return n * proc_factorial(n - 1)

print(proc_factorial(5))
````
