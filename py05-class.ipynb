````python
class FourCal:

  def __init__(self, a, b):#매개변수
    print('init')
    self.a = a #멤버변수
    self.b = b

  def add(self):
    return self.a + self.b  #메소드에서 멤버변수 사용가능

  def sub(self):
    return self.a - self.b

  def mul(self):
    return self.a * self.b
  
  def div(self):
    return self.a // self.b
````

### staticmethod
- 첫 번째 인자로 self, cls를 받지 않는다.
- 클래스나 인스턴스와 관계없는 일반적인 유틸리티 함수를 정의할때 사용한다.
- 독립적인 기능이지만 클래스 안에 넣는 것이 구조적으로 더 나을때 사용한다.

| 항목         | classmethod             | staticmethod |
| ---------- | ----------------------- | ------------ |
| 첫 번째 인자    | `cls` (클래스)             | 없음           |
| 클래스 변수 접근  | 가능                      | 불가능          |
| 인스턴스 변수 접근 | 불가능                     | 불가능          |
| 주 용도       | 클래스 상태를 사용하는 메서드        | 유틸리티 함수 정의   |
| 호출 방식      | 클래스.메서드() 또는 인스턴스.메서드() | 동일           |


````python
class B:
  x = 1


  def xx(self, x): # 인스턴스 메소드
    self.x = x


  @classmethod
  def yy(cls, x):  # 클래스 메소드
    cls.x = x


  @staticmethod
  def zz(a, b):  # 스테이틱 메소드
    print(a+b)


  @staticmethod  # 스테이틱 메소드
  def kk():
    print(x)
````

````python
B.yy(10)
````

````python
B.x
````

````python
B.zz(3,4)
````

````python
B.kk()
````

파이썬 입문의 저자는 박북스이고 200페이지로 구성됨


````python
class Book:
    
    def __init__(self, title=None, author=None, page=None):
        self.__title = title
        self.__author = author
        self.__page = page
    
    def getTitle(self):
        return self.__title
    
    def setTitle(self, title):
        self.__title = title

    def getAuthor(self):
        return self.__author
    
    def setAuthor(self, author):
        self.__author = author

    def getPage(self):
        return self.__page

    def setPage(self, page):
        self.__page = page

    def toString(self):
        return f"{self.__title}의 저자는 {self.__author}이고 {self.__page}페이지로 구성됨"
````

````python
class BookRun:
    def process(self):
        bk1 = Book()
        bk1.setTitle("파이썬의 정석")
        bk1.setAuthor('김파이썬')
        bk1.setPage(150)
        print(bk1.toString())
        #bk1.__title
        print(dir(bk1))
        # print(bk1._Book__title) #가지고 올 수 있다. 하지만 클래스이름이 붙음  #지양하는게 좋음!!!!!

        bk2 = Book('파이썬 입문', '박북스', 200)
        #하나씩 프린트 할 때는 getter쓰는게 좋음
        print('책이름 : {}'.format(bk2.getTitle()))
        print('책저자 : {}'.format(bk2.getAuthor()))
        print('페이지 : {}'.format(bk2.getPage()))
        print(bk2.toString())


if __name__ == '__main__':
    bn = BookRun()
    bn.process()
````

````python
class Dog:
    count = 0

    def __init__(self, name):
        self.name = name
        Dog.count = Dog.count + 1

    @classmethod
    def get_count(cls):
        return cls.count
````

정적 메소드로 유효한 이름인지 확인

````python
class Dog:
    count = 0

    def __init__(self, name):
        self.name = name
        Dog.count = Dog.count + 1

    @classmethod
    def get_count(cls):
        return cls.count
    
    @staticmethod
    def is_valid_name(name):
        #name의 클래스가 string이고 길이는 0보다 커야한다.
        return isinstance(name, str) and len(name) > 0
````

````python
#이름 유효성 검사는 객체와는 관련이 없음, 체크 같은거는 정적메소드 이용
print(Dog.is_valid_name('초코'))  #True
print(Dog.is_valid_name(''))     #False
print(Dog.is_valid_name(123))   #False
````

 - class에서 이름 뒤에 괄호는 상속이다.

 - python3부터는 (object)는 생략 가능하다. 즉 지금까지 class를 정의할때 object를 내부적으로 상속하고 있었다.

````python
# 부모클래스
class Parent:
  x = 'parent x'  #클래스 변수
  y = 'parent y'  #클래스 변수

  @classmethod
  def output(cls):
    return cls.x

  def test(self):
     return 10

  
#자식클래스
class Child(Parent):
  x = 'child x' # 클래스 변수

  def display(self):
    print('display')
    print(Parent.output())
    print(Child.output())
    print(Parent.test(self))
    print(Child.test(self))
````

````python
ch = Child()
print(Child.x)
print(Child.y)
````

````python
c = Child()
c.display()
````

# overriding(오버라이딩)
부모가 가진 기능을 변경하는 것을 overriding이라고 한다.  
double underbar 같 분은 것들은 overriding해서 객체의 기능을 바꿀 수 없다.


````python
# 부모 클래스 정의
class Animal:
    def __init__(self, name):
        self.name = name


    def make_sound(self):
        # 자식 클래스들이 반드시 재정의(오버라이딩)하도록 유도하는 부모의 기본 메서드
        return "알 수 없는 동물의 소리입니다."


# 자식 클래스 1 (Dog)
class Dog(Animal):
    def make_sound(self):
        # 부모의 make_sound 메서드를 자식에 맞게 완전히 새롭게 덮어씀 (오버라이딩)
        return "멍멍!"


# 자식 클래스 2 (Cat)
class Cat(Animal):
    def make_sound(self):
        # 부모의 make_sound 메서드를 자식에 맞게 완전히 새롭게 덮어씀 (오버라이딩)
        return "야옹~"
````

````python
animal = Animal("동물")
print(animal.make_sound())

dog1 = Dog("강아지이름")
print(dog1.make_sound())

cat1 = Cat("고냠미")
print(cat1.make_sound())
````

````python
from abc import ABC, abstractmethod

class Shape(ABC):

    def __init__(self, width, height):
        self.width = width
        self.height = height

    @abstractmethod
    def area(self):
      pass    

class Rect(Shape):
   
    def __init__(self, width, height ):
        super().__init__(width, height)

    def area(self):
      return self.width * self.height  
   

class Triangle(Shape):
   
    def __init__(self, width, height ):
        super().__init__(width, height)

    def area(self):
      return self.width * self.height * 0.5
````

````python
s = Shape(3, 2)
````

````python
r = Rect(2, 5)
print(f'사각형 넓이: {r.area()}')

t = Triangle(2, 5)
print(f"사각형 넓이: {t.area()}")
````
