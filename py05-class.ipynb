````python
class Person:
    def __init__(self, name, age, job):
      self.name = name
      self.age = age
      self.job = job


    def eat(self):
        print('eat')


    def walk(self):
        print('walk')
````

````python
class Hong(Person):  
    def study(self):
        print('study')
````

````python

hong = Hong('홍길동',30,'학생')
hong.eat()
hong.walk()
print(hong.name, hong.age, hong.job)
````

````python
class FourCal:

  def __init__(self, a, b):#매개변수
    print('init')
    self.a = a #멤버변수
    self.b = b

  def add(self):
    return self.a + self.b  #메소드에서 멤버변수 사용가능

  def sub(self):
    return self.a - self.b

  def mul(self):
    return self.a * self.b
  
  def div(self):
    return self.a // self.b
````

````python
fc = FourCal(4, 2) #생성자는 객체의 초기값을 할당한다.
````

````python
print(fc.add())
print(fc.sub())
print(fc.mul())
print(fc.div())
````

### decorator

````python
def t(a):
    return a
````

````python
t(3)
````

````python
def x(fun):
    def y(x):
        return fun(x+1)
    return y
````

````python
m = x(t)
print(m.__name__)
print(type(m))
m(5)
````

````python
x(t)(5)
````

````python
def x(fun):
    def y(x):
        return fun(x+1)
    return y
````

````python
#데코레이터
@x
#s = x(s)
def s(a):
    return a
````

````python
#s()함수에 @x 데코레이터가 선언되어 있지 않을 때 호출하는 방법
x(s)(5)

#s() 함수에 @x 데코레이터가 선언되어 있을 때 호출하는 방법
s(5)
````

### classmethod
- 첫 번째 인자로 클래스(cls)를 받는다.
- 클래스를 이용해 무언가를 해야할때 사용한다.
- 클래스 변수에 접근하거나 다른 클래스 메소드를 호출할 수 있다.
- 일반적으로 팩토리 메서드 등에 자주 사용된다.

````python
class B:
  x=3          # 클래스 변수
  @classmethod
  def aa(cls):  # classmethod는 관례상 cls를 사용한다.
    print('python')


  def bb(self):
    print('instance')
````

````python
B.x
````

````python
B.aa() #클래스.클래스메소드
````

````python
B.bb() #self가 받을 수 있는 값 넘겨야 함
````

````python
ab = B()
B.bb(ab)
````

````python
ab.aa()
````

````python
class B:
    x = 1

@classmethod
def aa(cls, x):
    cls.x = x
````

````python
B.x
````

````python
b.x #b인스턴스에 x가 없으므로 B클래스의 x를 가져온다.
````

````python
b.x = 5 # b인스턴스에 x변수를 생성한다.
````

````python
b.x
````

### staticmethod
- 첫 번째 인자로 self, cls를 받지 않는다.
- 클래스나 인스턴스와 관계없는 일반적인 유틸리티 함수를 정의할때 사용한다.
- 독립적인 기능이지만 클래스 안에 넣는 것이 구조적으로 더 나을때 사용한다.

| 항목         | classmethod             | staticmethod |
| ---------- | ----------------------- | ------------ |
| 첫 번째 인자    | `cls` (클래스)             | 없음           |
| 클래스 변수 접근  | 가능                      | 불가능          |
| 인스턴스 변수 접근 | 불가능                     | 불가능          |
| 주 용도       | 클래스 상태를 사용하는 메서드        | 유틸리티 함수 정의   |
| 호출 방식      | 클래스.메서드() 또는 인스턴스.메서드() | 동일           |


````python
class B:
  x = 1


  def xx(self, x): # 인스턴스 메소드
    self.x = x


  @classmethod
  def yy(cls, x):  # 클래스 메소드
    cls.x = x


  @staticmethod
  def zz(a, b):  # 스테이틱 메소드
    print(a+b)


  @staticmethod  # 스테이틱 메소드
  def kk():
    print(x)
````

````python
B.yy(10)
````
